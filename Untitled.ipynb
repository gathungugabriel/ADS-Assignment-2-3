# 1.Show a breakdown of distance from home by job role and attrition.

```python
data.columns
```

### Group data by job role and attrition

```python
grouped_data =  data.groupby(['JobRole','Attrition'])

```

### Calculate the average distance from home for each group


```python
avg_distance = grouped_data['DistanceFromHome'].mean().reset_index()
avg_distance
```

# 2. Compare average monthly income by education and attrition.

### Group data by Education and attrition

```python
grouped_educattrition = data.groupby(['Education','Attrition'])
```

```python
avg_mi = grouped_educattrition['MonthlyIncome'].mean().reset_index()
avg_mi
```

### Draw a nested barplot by job role and attrition

```python
  
sns.set_theme(style="whitegrid")



fig = sns.catplot(
    data=avg_distance, kind="bar",
    x="JobRole", y='DistanceFromHome', hue="Attrition",
     palette="dark", alpha=.6, height=6
)
fig.despine(left=True)
fig.set_axis_labels("JobRole", "Distance From Home")
plt.title("Distance from home by job role and attrition",loc="center",fontdict={'fontsize': 20})
fig.set_xticklabels(rotation=60)

plt.show()
```

Distance from home tends to be a factor contributing to attrition in six of the nine Job roles,
however the Manufacturing and Research Director as well as the Sales Representative roles defy this logic,
since the currently serving staff from these roles cover longer distance from home than those who already left.

### Draw a nested barplot by Education and attrition

```python
sns.set_theme(style="whitegrid")



fig = sns.catplot(
    data = avg_mi, kind="bar",
    x="Education", y='MonthlyIncome', hue="Attrition",
     palette="dark", alpha=.6, height=6
)
fig.despine(left=True)
fig.set_axis_labels("Education", "Monthly Income")
plt.title("Monthly Income by Education and attrition", loc = 'center',fontdict={'fontsize': 20})
plt.show()
```

In all education levels, Monthly Income is a factor affecting attrition whereby, all the staff who have left the company have had lower monthly income as compared to their colleagues within the same education level.
