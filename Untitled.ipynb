## Loading libraries

```python

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

## Loading dataset

```python
data=pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')
data.head(7)
```

```python

sns.set_theme(style="whitegrid")


# Draw a nested barplot by job role and attrition
fig = sns.catplot(
    data=data, kind="bar",
    x="Attrition", y="DistanceFromHome", hue="Department",
     palette="dark", alpha=.6, height=6
)
fig.despine(left=True)
fig.set_axis_labels("", "Distance From Home")
fig.legend.set_title("Distance from home by job role and attrition")
plt.show()
```
